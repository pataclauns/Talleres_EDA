#### Evaluar la necesidad de usar Label encoding y/o  hot-encoding

Para la pregunta SMART que se plantea, se quiere identificar del total de clientes que han ingresado cuantos estan inactivos en la Empresa, es decir ya no volvieron a comprar. Para este caso se usara la variable "Bloqueado", la cual es una variable cualitativa nominal, en ese orden par el ejercicio de llevar esta variable a un modelo, se podria aplicar 

```python
clientes_bloqueado= data.iloc[:,[0,10,14]].reset_index(drop=True)
clientes_bloqueado
```

```python
# se crea el objeto
onehotencoder = OneHotEncoder()
X = onehotencoder.fit_transform(clientes_bloqueado.Bloqueado.values.reshape(-1,1)).toarray()

X
```

```python
# Se agrega al dataset original
retirados = pd.DataFrame(X, columns = ["Bloqueado_"+str(int(i)) for i in range(X.shape[1])]) 
clientes_retirados = pd.concat([clientes_bloqueado, retirados], axis=1)
clientes_retirados
```

```python
# Calculamos Variance Inflation Factor(VIF).
# Cuantifica la  multicolinealidad en un análisis de regresión normal de mínimos cuadrados.

# Seleccionamos las variables que vamos a analizar 
X = clientes_retirados[['Cupo de crdito','Bloqueado_0', 'Bloqueado_1']]

# VIF dataframe
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns

# calculating VIF for each feature

vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
print(vif_data)
```

```python
#Calificación del VIF:
#VIF=1, Baja Multicollinearidad
#VIF<5, Moderada Multicollinearidad
#VIF>5, Multicollinearity Extrema (Se busca evitar)

#En este caso el VIF dio moderada multicollinearidad.
```
